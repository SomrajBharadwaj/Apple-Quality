```python
# Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Reading the dataset
df = pd.read_csv("D:/assignment/apple_quality2.csv")
```

```python
# Displaying the dataset
df
```

```python
# Dropping the row (4000)
df.drop(index=4000, inplace=True)
```

```python
# Display
df
```

```python
# Replacing good as 1 and bad as 0 
df['Quality'].replace({'bad': 0, 'good': 1}, inplace=True)
```

```python
# Display
df
```

```python
# Describe the dataset
print(df.describe())
```

```python
# Displaying the correlation of dataset 
print(df.corr())
```

```python
df.corr()
```

```python
# Plotting a Histogram

# Create a figure with a specified figure size
plt.figure(figsize=(10,6))

plt.hist(df['Sweetness'], bins=10, color='red', edgecolor= 'yellow')

# Add a title to the plot
plt.title('Histogram of Sweetness')

# Adding labels to the plot
plt.xlabel('Sweetness')
plt.ylabel('Count')

# Display
plt.show()
```

```python
# Plotting a Scatter graph

# Create a figure with a specified figure size
plt.figure(figsize=(10, 6))

plt.scatter(df['Weight'], df['Size'], alpha=0.5, color='green', edgecolors = 'w')

# Add a title to the plot
plt.title('Relationship Between Size and Weight')

# Adding labels to the plot
plt.xlabel('Size')
plt.ylabel('Weight')

# plotting the grid
plt.grid(True)

# Display
plt.show()
```

```python
# Function to create a Heatmap

# Create a figure with a specified figure size
plt.figure(figsize=(10, 6))

# Selecting the correlation matrix 
correlation_matrix = df.corr()

# Use seaborn function to plot correlation matrix in dataframe
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidth= 0.5)

# Add a title to the plot
plt.title('Heatmap of Correlation Matrix')

# Display
plt.show()
```
